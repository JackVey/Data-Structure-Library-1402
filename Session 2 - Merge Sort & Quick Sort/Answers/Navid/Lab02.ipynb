```python
import random
import time
import matplotlib.pyplot as plt
```

```python
# MergeSort in Python

def worstList(n, start = 100000):
    l = []
    for i in range(n):
       l.append(start - 5)
    return l
    
def bestList(n, start = 100000):
    l = []
    for i in range(n):
        l.append(start + 5)
    return l

def mergeSort(array):
    if len(array) > 50:

        #  r is the point where the array is divided into two subarrays
        r = len(array)//2
        L = array[:r]
        M = array[r:]

        # Sort the two halves
        mergeSort(L)
        mergeSort(M)

        i = j = k = 0

        # Until we reach either end of either L or M, pick larger among
        # elements L and M and place them in the correct position at A[p..r]
        while i < len(L) and j < len(M):
            if L[i] < M[j]:
                array[k] = L[i]
                i += 1
            else:
                array[k] = M[j]
                j += 1
            k += 1

        # When we run out of elements in either L or M,
        # pick up the remaining elements and put in A[p..r]
        while i < len(L):
            array[k] = L[i]
            i += 1
            k += 1

        while j < len(M):
            array[k] = M[j]
            j += 1
            k += 1

    else:
        array.sort()
```

```python
print("------------- Analysing BEST-CASE of MERGE-SORT -------------")
# Separate x and y values
x_values = []
y_values = []
for i in range(10):
    n = 100000 + 100000 * i
    worst_list = worstList(n)
    copy_of_list = worst_list.copy()
    st = time.process_time()
    mergeSort(worst_list)
    et = time.process_time()
    res = et - st
    print(f"time taken to sort {n} integers: {res} Seconds using merge sort in worst-case scenario, And sorting was ", end = "")
    arr = sorted(copy_of_list)
    x_values.append(n)
    y_values.append(res)
    if worst_list == arr:
        print("sucessful")
    else:
        print("failed") 
        
# Create the plot
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Add labels and a title
plt.xlabel('n-size of input')
plt.ylabel('time taken(in seconds)')
plt.title('Running time for random-case scenario of Insertion Sort')

# Show the plot
plt.show()
```
